# PREGUNTA 1 - 60 PUNTOS

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit, execute,IBMQ
anc = QuantumRegister(1, 'scratch')
q = QuantumRegister(6,'q')
c = ClassicalRegister(6,'c')
circuitp1 = QuantumCircuit(q,anc,c)
circuitp1.x(anc[0])
circuitp1.h(anc[0])
circuitp1.h(q[0:6])
circuitp1.cx(q[0:6],anc[0])
circuitp1.barrier(q)
circuitp1.h(q[0:6])
circuitp1.measure(q,c)

circuitp1.draw()
```
